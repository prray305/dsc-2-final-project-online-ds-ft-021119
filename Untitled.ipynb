___Random sampling distribution___

```python
def get_sample(data, n=30):
    sample = []
    while len(sample) != n:
        rand = np.random.choice(data)
        sample.append(rand)
    return sample
```

```python
def get_sample_mean(sample):
    return np.mean(sample)
```

```python
def create_sample_distribution(data, dist_size=1000, n=30):
    sample_dist = []
    while len(sample_dist) != dist_size:
        sample = get_sample(data, n)
        sample_mean = get_sample_mean(sample)
        sample_dist.append(sample_mean)
    
    return sample_dist
```

___Perform t-test___

Since the variances are not equal, Welch's t-test will be best approach to calculate the critical value.

___Calculate the t-statisitic___

```python
def welch_t(sample1, sample2):
    mean1 = np.mean(sample1)
    mean2 = np.mean(sample2)
    numerator = mean1 - mean2
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)
    n1 = len(sample1)
    n2 = len(sample2)
    denominator = np.sqrt(var1/n1 + var2/n2)
    return numerator / denominator
```

```python
print(np.mean(full_orders_dist))
print(np.mean(disc_orders_dist))
```

```python
t = welch_t(disc_orders_dist, full_orders_dist)
```

___Calculate the degrees of freedom___

```python
def df_welch(sample1, sample2):
    s1 = np.var(sample1, ddof=1)
    s2 = np.var(sample2, ddof=1)
    n1 = len(sample1)
    n2 = len(sample2)
    
    numerator = (s1/n1 + s2/n2)**2
    denominator = (s1/n1)**2/(n1-1) + (s2/n2)**2/(n2-1)
    
    return numerator / denominator
```

```python
df = df_welch(disc_orders_dist, full_orders_dist)
```

___Plot t-statistic and disctribution___

```python
fig, ax = plt.subplots()
x = np.linspace(-5, 5, 200)
t_dist = stats.t.pdf(x, df)
ax.plot(x, t_dist, label='t-distribution')
ax.vlines(t, 0, .4, linestyles='--', color='k', label='t-statistic = {:.3f}'.format(t))
ax.set_ylabel('pdf(t)')
ax.set_xlabel('t')
ax.set_ylim(0, 0.55)
ax.legend(loc=9)
ax.fill_between(x, 0, t_dist, where=(x >= t), color='grey', alpha=0.5)
```

```python
## confirm equation and result
p = 1 - stats.t.cdf(t, df)
p
```

```python
def p_value(sample1, sample2):
    t = welch_t(sample1, sample2)
    df = df_welch(sample1, sample2)
    
    p = stats.t.cdf(-np.abs(t), df)
    
    return "t-stat:{}".format(t), "p-value:", (p+p)
```

```python
p_value(disc_orders_dist, full_orders_dist)
```

___Measure the effect size.___

```python
def Cohen_d(sample1, sample2):
    diff = np.mean(sample1) - np.mean(sample2)
    
    n1, n2 = len(sample1), len(sample2)
    var1 = np.var(sample1)
    var2 = np.var(sample2)
    
    pooled_var = (n1 * var1 + n2 * var2) / (n1 + n2)
    d = diff / np.sqrt(pooled_var)
    
    return d
```

```python
Cohen_d(disc_orders_dist, full_orders_dist)
```

Per the Cohen's D test, the results, or effect size, of the significance test are not large but rather medium.

___Plot overlap and superiority of two groups___

```python
def evaluate_PDF(rv, x=4):

    mean=rv.mean()
    std =rv.std()
    
    xs = np.linspace(mean - x*std, mean + x*std, 100)
    ys = rv.pdf(xs)
    
    return xs, ys
```

```python
def overlap_superiority(group1, group2, n=1000):
    group1_sample = group1.rvs(n)
    group2_sample = group2.rvs(n)
    
    thresh = (group1.mean() + group2.mean()) / 2
    print(thresh)
    
    above = sum(group1_sample < thresh)
    below = sum(group2_sample > thresh)
    
    overlap = (above + below) / n
    
    superiority = sum(x > y for x, y in zip(group1_sample, group2_sample)) / n
    
    return overlap, superiority
```

```python
def plot_pdfs(cohen_d=1.6):
    group1 = scipy.stats.norm(0, 1)
    group2 = scipy.stats.norm(cohen_d, 1)
    xs, ys = evaluate_PDF(group1)
    plt.fill_between(xs, ys, label='Group1', alpha=0.7)
    
    xs, ys = evaluate_PDF(group2)
    plt.fill_between(xs, ys, label='Group2', alpha=0.7)
    
    o, s = overlap_superiority(group1, group2)
    print('overlap', o)
    print('superiority', s)
```

```python
#calculate what is required sample size for stat significance
control_mean = np.mean(full_orders_dist)
control_std = np.std(full_orders_dist)
experimental_mean = np.mean(disc_orders_dist)
experimental_std = np.std(disc_orders_dist)
sample_size = 30
```

```python
target = .95
```

```python
n = 1
n_sim = 10000
current = 0
```

```python
p = np.empty(n_sim)
p.fill(np.nan)
power_sample = []
while current < target:
    data = np.empty([n_sim, n, 2])
    data.fill(np.nan)
    
    data[:, :, 0] = np.random.normal(loc= control_mean, scale=control_std, size=[n_sim, n])
    
    data[:, :, 1] = np.random.normal(loc= experimental_mean, scale=experimental_std, size=[n_sim, n])
#     data[:, :, 0] = np.random.normal(loc= 100, scale=10, size=[n_sim, n])
    
#     data[:, :, 1] = np.random.normal(loc= 50, scale=20, size=[n_sim, n])
    
    result = stats.ttest_ind(data[:, :, 0], data[:, :, 1], axis=1)
    p = result[1]
    
    rejects = np.sum(p< 0.05)
    reject_proportion = rejects / float(n_sim)
    current = reject_proportion
    
    print ("Sample size:", n, ", Calculated power: ", current)
    power_sample.append([n, current])
    n +=1
```
